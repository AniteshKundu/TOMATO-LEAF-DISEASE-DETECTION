```python
import os
import csv
import shutil

# Path to the directory containing the images
image_folder = '/content/drive/MyDrive/new proj/Tomato_images'

# Path to the CSV file
csv_file = '/content/drive/MyDrive/new proj/train.csv'

# Directory to move or copy the matching images
output_folder = '/content/drive/MyDrive/new proj/IMAGE'

# Read the CSV file and extract the image names
image_names = []
with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        image_names.append(row['img'])  # Replace 'name_column' with the actual column name in your CSV

# Get a list of all image files in the folder
all_images = os.listdir(image_folder)

# Find matching images and move/copy them to the output folder
for image_name in image_names:
    if image_name in all_images:
        image_path = os.path.join(image_folder, image_name)
        output_path = os.path.join(output_folder, image_name)
        shutil.copy(image_path, output_path)  # Change copy to move if you want to move the files

print("Images copied/moved successfully.")
```
